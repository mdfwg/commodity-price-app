```python
import pandas as pd

# Load Dataset
df1 = pd.read_csv("stagnan_filled.csv")
df2 = pd.read_csv("wavelet_volatile_filled.csv")

# Gabungkan dataset
df = pd.concat([df1, df2], ignore_index=True)
df
```

```python
df_result = pd.read_csv("outputfinal.csv")
df_result
```

```python
import pandas as pd

# Step 1: Split 'id' column in df_result
df_result[['Product', 'Region', 'Date']] = df_result['id'].str.split('/', expand=True)

# Step 2: Convert 'Date' to datetime and rename 'price' to 'Price'
df_result['Date'] = pd.to_datetime(df_result['Date'])
df_result = df_result.rename(columns={'price': 'Price'})

# Step 3: Reorder columns
df_result = df_result[['Date', 'Region', 'Price', 'Product']]

# Step 4: Ensure df['Date'] is datetime
df['Date'] = pd.to_datetime(df['Date'])

# Step 5: Concatenate
df_combined = pd.concat([df, df_result], ignore_index=True)

# Step 6: Convert 'Date' to just date
df_combined['Date'] = df_combined['Date'].dt.date

# ✅ Step 7: Sort by Date → Region → Product
df_combined = df_combined.sort_values(by=['Date', 'Region', 'Product']).reset_index(drop=True)

# Optional: Format float for display
pd.set_option('display.float_format', '{:,.6f}'.format)

# Display result
print(df_combined.head())
```

```python
import pandas as pd
from datetime import datetime

# Assume: df_combined, region_map, commodity_map, user_id are already defined
output_data = []

for index, row in df_combined.iterrows():
    # Clean and parse the date
    if isinstance(row["Date"], str):
        date = datetime.strptime(row["Date"], "%Y-%m-%d").date()
    elif isinstance(row["Date"], datetime):
        date = row["Date"].date()
    else:
        date = row["Date"]

    try:
        region_id = region_map[row["Region"].strip()]
        commodity_id = commodity_map[row["Product"].strip()]
        price = float(str(row["Price"]).replace(",", ""))

        output_data.append({
            "region_id": region_id,
            "commodity_id": commodity_id,
            "date": date.isoformat(),
            "price": price,
            "created_by": user_id  # Should be the actual UUID of the admin
        })
    except Exception as e:
        print(f"[SKIPPED row {index}] {e}")

# Convert to DataFrame
df_output = pd.DataFrame(output_data)

# Save as CSV (without index, for clean import)
df_output.to_csv("bulk_prices_upload.csv", index=False)
```
